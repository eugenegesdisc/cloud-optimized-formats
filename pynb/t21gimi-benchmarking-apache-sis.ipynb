## Set Environmental Variables


```python
import os
from pathlib import Path
home_directory = Path.home()

print(f"User's home directory is: {home_directory}")

default_data_dir = os.path.join(home_directory, "work","data","t21","data4t21")
data_dir = input(f"Enter your data directory (default: {default_data_dir}): ") or default_data_dir
print(f"You entered data directory: {data_dir}")

default_prog_dir = os.path.join(home_directory, "work","wksp","java","EDIT-T21-GIMI", "D213", "code")
prog_dir = input(f"Enter your program directory (default: {default_prog_dir}): ") or default_prog_dir
print(f"You entered program directory: {prog_dir}")
```

## Generate Scenarios

```python
import os
import subprocess

BUILD_DIR = os.path.join(default_prog_dir, "target")
print(f"BUILD_DIR={BUILD_DIR}")

bm_args = [
    "java", f"-Dderby.stream.error.file=\"{BUILD_DIR}/derby.log\"",
    "--enable-native-access=org.apache.sis.storage.gdal",
    "--add-exports", "org.apache.sis.storage/org.apache.sis.storage.base=org.apache.sis.storage.geoheif",
    "--add-exports", "org.apache.sis.storage/org.apache.sis.io.stream=org.apache.sis.storage.geoheif",
    "--add-exports", "org.apache.sis.util/org.apache.sis.util.internal.shared=org.apache.sis.storage.geoheif",
    "--add-exports", "org.apache.sis.feature/org.apache.sis.image.internal.shared=org.apache.sis.storage.geoheif",
    "--add-exports", "org.apache.sis.util/org.apache.sis.pending.jdk=org.apache.sis.storage.geoheif",
    "--module-path", f"{BUILD_DIR}/dependency:{BUILD_DIR}/T21-D213-1.0-SNAPSHOT.jar",
    "--module", "org.opengis.testbed.gimi/org.opengis.testbed.T21D213.Command",
    "generate-scenarios",
    "--image-directory",
    f"{data_dir}",
    "--images",
    "COMPARE_COG,COMPARE_HEIF",
    "--scenarios",
    "SERIAL",
]

f_cmd = " ".join(bm_args)
print(f_cmd)
# Run the compiled Java class with an argument
try:
    result = subprocess.run(bm_args, capture_output=True, text=True, check=True)

    print("SUCCESS-STDOUT:", result.stdout)
    print("SUCCESS-STDERR:", result.stderr)
except subprocess.CalledProcessError as e:
    print(f"Subprocess failed with return code {e.returncode}")
    print("STDOUT:", e.stdout)
    print("STDERR:", e.stderr)
except FileNotFoundError:
    print("Error: Java executable not found. Check your PATH environment variable.")
```

## Running the Benchmarking Test



```python
import os
import subprocess

BUILD_DIR = os.path.join(default_prog_dir, "target")
print(f"BUILD_DIR={BUILD_DIR}")

bm_args = [
    "java", f"-Dderby.stream.error.file=\"{BUILD_DIR}/derby.log\"",
    "--enable-native-access=org.apache.sis.storage.gdal",
    "--add-exports", "org.apache.sis.storage/org.apache.sis.storage.base=org.apache.sis.storage.geoheif",
    "--add-exports", "org.apache.sis.storage/org.apache.sis.io.stream=org.apache.sis.storage.geoheif",
    "--add-exports", "org.apache.sis.util/org.apache.sis.util.internal.shared=org.apache.sis.storage.geoheif",
    "--add-exports", "org.apache.sis.feature/org.apache.sis.image.internal.shared=org.apache.sis.storage.geoheif",
    "--add-exports", "org.apache.sis.util/org.apache.sis.pending.jdk=org.apache.sis.storage.geoheif",
    "--module-path", f"{BUILD_DIR}/dependency:{BUILD_DIR}/T21-D213-1.0-SNAPSHOT.jar",
    "--module", "org.opengis.testbed.gimi/org.opengis.testbed.T21D213.Command",
    "benchmark",
    "--image-directory",
    f"{data_dir}",
    "--images",
    "COMPARE_COG,COMPARE_HEIF",
    "--scenarios",
    "SERIAL",
    "--num-passes", "110",
    "--ignored-passes", "10"
]

f_cmd = " ".join(bm_args)
print(f_cmd)
# Run the compiled Java class with an argument
try:
    result = subprocess.run(bm_args, capture_output=True, text=True, check=True)

    print("SUCCESS-STDOUT:", result.stdout)
    print("SUCCESS-STDERR:", result.stderr)
except subprocess.CalledProcessError as e:
    print(f"Subprocess failed with return code {e.returncode}")
    print("STDOUT:", e.stdout)
    print("STDERR:", e.stderr)
except FileNotFoundError:
    print("Error: Java executable not found. Check your PATH environment variable.")
```
